```python
import torch
import reid.losses
import matplotlib.pyplot as plt
import numpy as np
```

```python
criterion = reid.losses.ContrastiveLoss()
```

```python
def get_loss(x):
    return criterion(torch.Tensor([[0]]), torch.Tensor([[x]]), 0)
```

```python
x = np.arange(-1, 1, 0.01)
plt.plot(x, [get_loss(x_) for x_ in x])
```
